Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

ANS:
five common plots that you can create using the Seaborn library along with their typical use cases:

1. **Scatter Plot**:
   - **Use**: Visualizing the relationship between two continuous variables. It helps in identifying patterns, clusters, or trends in data points.
   
2. **Line Plot**:
   - **Use**: Showing the trend of a continuous variable over a continuous interval or time series data. It's useful for visualizing data with sequential order.
   
3. **Histogram**:
   - **Use**: Displaying the distribution of a single numerical variable. It helps in understanding the frequency or density of observations within different bins or intervals.
   
4. **Bar Plot**:
   - **Use**: Comparing the values of different categories. It's suitable for showing and comparing the counts, sums, or averages of categorical data.
   
5. **Heatmap**:
   - **Use**: Visualizing the correlation or relationship between multiple variables. It's particularly useful for exploring patterns and identifying relationships in large datasets.


Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions. 
Note:  timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# load the fmri
dataset=sns.load_dataset("fmri")
# line plot
plt.figure(figsize=(10,6)) #width= 10 inches , height= 6 inches
sns.lineplot(data=dataset,x ="timepoint", y = "signal")
# title and labels
plt.title("FMRI SIGNAL over TIME BY EVENT AND REGION ")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# show
plt.legend(title='EVENT')
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

Note:  pclass, age, and fare are columns in the titanic dataset.


```python
# load the fmri
dataset=sns.load_dataset("titanic")
# line plot
plt.figure(figsize=(12,6)) #width= 12 inches , height= 6 inches

# box plot for age
plt.subplot(121)
sns.boxplot(data=dataset,x ='pclass', y = 'age')
plt.title("AGE DISTIBUTION BY PASSENGER CLASS")
plt.xlabel("PASSENGER CLASS")
plt.ylabel("AGE")


# box plot for age
plt.subplot(122)
sns.boxplot(data=dataset,x ='pclass', y = 'fare')
plt.title("FARE BY PASSENGER CLASS")
plt.xlabel("PASSENGER CLASS")
plt.ylabel("FARE")

# show

plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset. 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# load the fmri
dataset=sns.load_dataset("diamonds")
# line plot
plt.figure(figsize=(10,6)) #width= 10 inches , height= 6 inches
sns.histplot(data=dataset,x ="price", hue='cut')
# title and labels
plt.title("DISTRIBUTION OF DIAMOND BY CUT ")
plt.xlabel("PRICE")

# show

plt.show()
```


Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset. 


```python
import seaborn as sns
import matplotlib.pyplot as plt

# load the fmri
dataset=sns.load_dataset("iris")
# line plot
plt.figure(figsize=(10,6)) #width= 10 inches , height= 6 inches
sns.pairplot(data=dataset, hue='species')
# title and labels
plt.title("IRIS DISTRIBUTION ")


# show

plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# load the fmri
dataset=sns.load_dataset("flights")

# matrix for heatmap
flight_pivot=dataset.pivot("month","year","passengers")

# line plot
plt.figure(figsize=(10,6)) #width= 10 inches , height= 6 inches
sns.heatmap(data=flight_pivot,annot=True, fmt="d", cmap="YlGnBu")
# title and labels

plt.title("flights DISTRIBUTION ")


# show
plt.show()
```
